### Descripción

NSL-KDD is a data set suggested to solve some of the inherent problems of the KDD'99 data set which are mentioned in [1]. Although, this new version of the KDD data set still suffers from some of the problems discussed by McHugh [2] and may not be a perfect representative of existing real networks, because of the lack of public data sets for network-based IDSs, we believe it still can be applied as an effective benchmark data set to help researchers compare different intrusion detection methods. Furthermore, the number of records in the NSL-KDD train and test sets are reasonable. This advantage makes it affordable to run the experiments on the complete set without the need to randomly select a small portion. Consequently, evaluation results of different research work will be consistent and comparable.

## Funciones Auxiliares

```python
def load_kdd_dataset(data_path): # Función creada anteriormente 
    """Lectura del conjunto de datos NSL-KDD."""
    with open(data_path, 'r') as train_set:
        dataset = arff.load(train_set)
    attributes = [attr[0] for attr in dataset["attributes"]]
    return pd.DataFrame(dataset["data"], columns=attributes)
```

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None): # Revuelve el particionado de datos y hace la selección 
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split( # Se toma el train set y el train test  
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat) # Se testea con el 40% y se entrena con el 60%
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

## API de SKLEARN

Antes de continuar, vamos a hacer una reseña de como funcionan las API´s de SKLEAR

* **Estimators:** Cualquiier objeto que pueda estimar algun parámetro:
    * El propio estimador, se forma mediante el método fit() que siempre toma un DataSet como argumento.
    *  Cualquier otro parámetro de este método, es un hiperparámetro
*  **Transformers:** Son estimadores capaces de  transformar el DataSet (como imputer) .
    *  La transformación se realiza mediante el metodo transform().
    *  Reciben un DataSet como parámetro de entrada.
*  **Predictors:** SOn estimadores capaces de realizar predicciones.
    *   La predicción se realiza mediante el método predict().
    *   Reciben un DataSet como entrada.
    *   Retorna un DataSet con las predicciones.
    *   Tiene un método score() para evaluar el resultado de la predicción.

## Transformadores para atributos numéricos

```python
# Transformador creado para eliminar las filas con valores Nulos
from sklearn.base import BaseEstimator, TransformerMixin

class DeleteNanRows(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        return X.dropna()


    
```

```python
# Transformador diseñado para escalar de manera sencilla únicamente algunas columnas seleccioadas
class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, attributes):
        self.attributes = attributes
    def fit(self, X, y = None):
        return self # Nothing else to do
    def transform(self, X, y = None):
        X_copy = X.copy()
        scale_attr = X_copy[self.attributes]
        robust_scaler = RobustScaler() 
        # ROBUST_SCALER Iguala los datos para que tengan la misma escala
        # Reduce el impacto de  de datos extremos
        X_scaled = robust_scaler.fit_transform(scale_attr)
        X_scaled = pd.DataFrame(X_scaled, columns = self.attributes, index = X_copy.index) # Asignar una escala para que no afecte los datos atipicos y se convierte a un DataFRame
        for attr in self.attributes:
            X_copy[attr] = X_scaled[attr]
        return X_copy
```

## 6.- Construyendo Pipelines Personalizados

Los Pipelines permiten agrupar en un flujo de ejecución todas las operaciones de transformación que se necesitan realizar  sobre el DataSet, esto facilita muchisisimo  las transformaciones para diferentes DataSets. Los aspectos a tener en cuenta en estas estructuras son :
* Recibe un DataSet de pares (nombre, estimador)
* Todos, menos el último deben ser transformadores.
* El pipeline expone los mismos métodos que el **último estimador**.
* CUando se llama el método `fit()` del pipeline, se llama secuencialmente al método `fit_transform()` de los estimadores, y se les pasa de manera secuencial el output del siguiente. El último invoca al método `fit()`

```python
# Construcción de un pipeline para los atributos numéricos
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('rbst_scaler', RobustScaler()),
     ])
```

```python
# La clase imputer no admite valos categóricos, se eliminan los valores categóricos
X_train_num = X_train.select_dtypes(exclude=['object'])
X_train_prep = num_pipeline.fit_transform(X_train_num)
X_train_prep = pd.DataFrame(X_train_prep, columns = X_train_num.columns, index=X_train_num.index)
```

A continuación se presenta el método **ColumnTransformer**, que ejecuta todos los pipeline y concatena el resultado para ello el resultado de lo spipelines debe ser un valor numérico

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(X_train.select_dtypes(exclude = ['object']))
cat_attribs = list(X_train.select_dtypes(include = ['object']))

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
```

```python
X_train_prep = full_pipeline.fit_transform(X_train)
```

```python
X_train_prep = pd.DataFrame(X_train_prep, columns = list(pd.get_dummies(X_train)), index = X_train.index)
```
